```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import datetime 
import pickle 
```

```python
df = pd.read_csv("./data/Toy_data.csv")
```

```python
df.rename(columns = {'Unnamed: 0':'Date'}, inplace = True)
```

```python
df.shape
```

```python
# Date control 

df['Date'] = df.Date.apply(lambda x : datetime.datetime.strptime(x, '%Y-%m-%d') )
df = df.set_index('Date')
```

```python
with open("./bench_dict.pkl", 'rb') as f:
    bench_dict = pickle.load(f)
```

```python
past_duration = 400

st_ind = past_duration//30


last_t = bench_dict[st_ind-1][1]

past_days = 90

first_t = last_t - datetime.timedelta(days = past_days)
df_t = df[(df.index<=last_t) & (df.index>=first_t)]

X_input = []
past_days = 90
for i in range(st_ind, len(bench_dict)):
    last_t = bench_dict[i-1][1]
    first_t = last_t - datetime.timedelta(days = past_days)
    df_t = df[(df.index <= last_t) & (df.index >= first_t)]
    
    X_input.append(df_t)
```
